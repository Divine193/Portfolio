## Data Cleaning

```python
# Remove sensitive/unnecessary columns
sensitive_cols = ['TRANSACTION_ID', 'CUSTOMER_ID', 'POLICY_NUMBER', 'ADDRESS_LINE1', 'ADDRESS_LINE2', 'CITY', 'SSN', 'ROUTING_NUMBER','INCIDENT_CITY', 'ACCT_NUMBER', 'VENDOR_ID', 'POSTAL_CODE', 'CUSTOMER_NAME', 'AGENT_ID']
existing_data = insure_data.drop(sensitive_cols, axis=1)
```

```python
# Check the existing data
existing_data.info()
```

### Handle Missing Values

```python
# Check for missing values
existing_data.isna().sum()
```

### Filling the missing values of the Customer_Education_Level with "Unknown" because it is an important part of our analysis and it will alter the Insuance_Type column if dropped

```python
# fill the missing values with unknown
existing_data['CUSTOMER_EDUCATION_LEVEL'] = existing_data['CUSTOMER_EDUCATION_LEVEL'].fillna('Unknown')
```

```python
# Quick Check
existing_data.info()
```

### Finding Duplicates

```python
existing_data.duplicated().sum()
```

### Renaming Columns 

```python
# Convert the columns from uppercase to a proper snake_case
existing_data.columns = existing_data.columns.str.strip().str.lower()
```

```python
# Quick Check
existing_data.info()
```

Changing the "tenure" column to ""policy_tenure_mnths" which indicates the tenure in months

```python
existing_data.rename(columns = {'tenure' : 'policy_tenure_mnths'}, inplace=True)
```

```python
# Quick Check
existing_data.info()
```

```python
# convert the date columns to datetime datatype and check 
date_columns = ['txn_date_time', 'loss_dt', 'policy_eff_dt', 'report_dt']

for col in date_columns:
    existing_data[col] = pd.to_datetime(existing_data[col], errors = 'coerce')
existing_data.info()
```

### Feature Engineering

```python
# Create two new columns for transaction delay and reporting delay
existing_data['txn_delay'] = existing_data['txn_date_time'] - existing_data['report_dt']
existing_data['reporting_delay'] = existing_data['report_dt'] - existing_data['loss_dt']
existing_data.info()
```

```python
# Extract the day section from the created columns so it can be an integer
existing_data['reporting_delay'] = existing_data['reporting_delay'].dt.days
existing_data['txn_delay'] = existing_data['txn_delay'].dt.days
existing_data.info()
```
